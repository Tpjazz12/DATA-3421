## 4. Variability Analysis

```python
description = data.describe()
print(description.loc['std'])
```

* Budget_M has the highest standard deviation by a considerable margin, which indicates that it has the highest variability among the variables listed. This aligns with the histogram plots which showed a wide spread of values and a right-skewed distribution.
* Audience Ratings also has a high standard deviation, suggesting significant variability. However, without the full context of the maximum and minimum values, it's hard to compare this directly with Budget_M. Given the values we have, Budget_M still has a higher standard deviation.
* Rotten Tomatoes and IMDB have much lower standard deviations, indicating less variability in their values compared to Audience Ratings and Budget_M.
* Therefore, Budget_M demonstrates the highest variability.

## 5. Correlation Analysis 

```python
# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=[np.number])
```

```python
correlation_matrix = numeric_data.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

#### Positive Correlation with Sales Revenue_M:
Budget_M shows a positive correlation coefficient with Sales Revenue_M (approximately 0.03). This indicates a positive correlation, but the value is very close to zero, suggesting that the correlation is very weak.
#### Variables with Strong Correlation:
In the context of this heatmap, none of the variables display a strong correlation with Sales Revenue_M.The highest positive correlation coefficient visible in the heatmap with Sales Revenue_M is with Budget_M, but as mentioned, it's a very weak positive correlation
#### Higher Budget and Higher Sales Revenue:
The correlation coefficient between Budget_M and Sales Revenue_M does not indicate a strong relationship; thus, based on this matrix, we cannot conclude that a higher budget correlates strongly with higher sales revenue. However, the positive value does suggest that, in general, as the budget increases, there might be a tendency for sales revenue to increase as well, even if the relationship is not strong.

## 6. Average Profit by Genre

```python
# The average sales revenue for each genre
average_revenue_by_genre = data.groupby('Genre')['Sales Revenue_M'].mean()
average_revenue_by_genre
```

```python
# Sort the genres by the calculated mean sales revenue in descending order
sorted_revenue_by_genre = average_revenue_by_genre.sort_values(ascending=False)
sorted_revenue_by_genre
```

```python
# Display the genre with the highest average sales revenue
highest_revenue_genre = sorted_revenue_by_genre.idxmax()
highest_revenue_value = sorted_revenue_by_genre.max()
```

```python
print(f"The genre with the highest average sales revenue is {highest_revenue_genre} with an average revenue of {highest_revenue_value}.")
```

## 7. Marketing Spend  Analysis 

```python
correlation = data['Budget_M'].corr(data['Sales Revenue_M'])
print(f"The correlation between budget and sales revenue is: {correlation}")
```

```python
# A Scatter plot to visualize the relationship
sns.scatterplot(x='Budget_M', y='Sales Revenue_M', data=data)
plt.title(' Budget vs Sales Revenue')
plt.xlabel('Budget (M$)')
plt.ylabel('Sales Revenue (M$)')
plt.show()
```

* The correlation between Budget and Sales revenue is approximately 0.03. This indicates a very weak positive correlation between the two variables.
* The scatter plot also seems to show that there is no clear upward trend that would indicate a strong positive relationship; the data points do not form a pattern that rises to the right. Instead, they are quite spread out, with some high sales revenues occurring at various budget levels, including some lower-budget films.
* Based on the  correlation coefficient and the scatter plot, a higher budget does not strongly correlate with higher sales revenue in this dataset.

## 10. Noisy Data Analysis 

```python
###Check for Invalid Entries for numerical columns

invalid_entries = data[data.select_dtypes(include=['number'])<0].count()
print("Invalid Entries:\n", invalid_entries)
```

* The output indicates that there are no such entries across all the numerical columns in the dataset, which suggests that, at least in terms of negative values, the data does not contain these specific types of errors or noise.

* The effects of noisy data on analysis can include:
  * Skewed analysis and statistical measures (mean, median, variance, etc.)
  * Poor model performance for predictive modeling or machine learning
  * Misleading results leading to incorrect conclusions
