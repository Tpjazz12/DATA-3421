```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
######### Understanding the distribution of categorical variables

## Create a for loop to count the values for the categorical variables

for col in df.select_dtypes(include='object').columns:
  print(f"Value counts for {col}:\n{df[col].value_counts()}\n")
```

# **Check the data Types**

```python
data_types=df.dtypes
print(data_types)
```

```python
### Convert "Date of Joining" variable to a datetime format

df["Date of Joining"]=pd.to_datetime(df["Date of Joining"])
```

```python
### Check the Date of Joining format

print(df["Date of Joining"].dtype)
```

```python
### Example: Convert the Age column to numeric

df["Age"]=pd.to_numeric(df["Age"], errors="coerce")
```

# **3. Check for Invalid Entries**

```python
###Check for Invalid Entries for numerical columns

invalid_entries=df[df.select_dtypes(include=['number'])<0].count()

print("Invalid Entries:\n", invalid_entries)
```

```python
#### Handling and Cleaning Invalid Entries


### Option 1. Handling Invalid Entries in "Age" columns, Replace the negative values with NaN

df.loc[df['Age']<0,'Age']=None
```

```python
### Option 2: Replace the negative values with the median/mean age (Uncommon to use)

median_age=df['Age'][df['Age']>=0].median()
df.loc[df['Age']<0,'Age']=median_age
```

```python
df.loc[df['Dept']<0,'Dept']=None
```

```python

invalid_entries=df[df.select_dtypes(include=['number'])<0].count()

print("Invalid Entries:\n", invalid_entries)
```

#**4. Check for Duplicate records**

```python
duplicates=df.duplicated().sum()
print("Number of Duplicate Records:",duplicates )
```

```python
### Remove the Duplicates

df=df.drop_duplicates()
```

# **5. Initial Data Visualization**

```python
#### Histograms and Density Plots for numerical Variables

for col in df.select_dtypes(include=['number']).columns:
  sns.histplot(df[col], kde=True)
  plt.title(f"Histogram and Debisty plot of {col}")
  plt.show()
```

```python
### Create boxplots

for col in df.select_dtypes(include=['number']).columns:
  sns.boxplot(x=df[col])
  plt.title(f"Box Plot of {col}")
  plt.show()
```

```python
### Scatter plot for understanding the relationships between numerical variables

sns.pairplot(df.select_dtypes(include=['number']), kind='reg')
plt.suptitle("Paitwise Scatterplots of numerical variables")
plt.show()
```

### **Correlation Matrix**

```python
correlation_matrix=df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
